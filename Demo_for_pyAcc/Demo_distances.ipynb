```python
import sys
sys.path.append('/home/elena/astrocosmolib/')

import numpy as np
import matplotlib.pyplot as plt

from astropy import units as u

from pyACC.cosmology.distances import CosmologicalDistances

import pandas as pd
```

```python
H0_planck = 67.4
Omega_m_planck = 0.315
z = np.linspace(0, 10, 100)
```

```python
def hubble_lcdm_func(z, Omega_M, H0):
    return H0 * np.sqrt(Omega_M * (1 + z)**3 + (1 - Omega_M))

dist_calculator = CosmologicalDistances(hubble_lcdm_func, Omega_m_planck, H0_planck)

dc = []
da = []
dl = []

for redshift in z:
    dc.append(dist_calculator.comoving_distance(redshift))
    da.append(dist_calculator.angular_distance(redshift))
    dl.append(dist_calculator.luminosity_distance(redshift))

print("Comoving distances:", dc)
print("Angular distances:", da)
print("Luminosity distances:", dl)
```
