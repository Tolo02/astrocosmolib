```python
import numpy as np
import matplotlib.pyplot as plt

import sys 
sys.path.append('/home/elena/astrocosmolib/')

from pyACC.clustering.measure.two_points import TwoPointIsotropic
```

```python
# Generate two sample of random points in a cube of size L
L = 1000

# Number of points
n = 100000

# First set of points
x1, y1, z1 = np.random.rand(3, n) * L

# Second set of points
x2, y2, z2 = np.random.rand(3, n) * L

# Create the set of points
twopcf = TwoPointIsotropic()

# Set the scales
r_min = 0
r_max = 50
delta_r = 1

# Compute the two-point correlation function
twopcf.compute(r_min, r_max, delta_r, x1, y1, z1, x2, y2, z2, periodic=True, boxsize=L)

# Plot the two-point correlation function
plt.plot(twopcf.r_centers, twopcf.xi, label="Two-point correlation function")
plt.xlabel(r"$r$ [Mpc/h]")
plt.ylabel(r"$\xi(r)$")
```
